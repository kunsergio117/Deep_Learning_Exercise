```python
import nltk
import spacy

#download the corpora
#nltk.download()

from nltk.corpus import gutenberg
```

```python
gutenberg.fileids()
```

```python
shakespeare = gutenberg.raw(fileids='shakespeare-caesar.txt')
print(shakespeare)
```

# Sentence Segmentation

This is known as sentence tokenizer in NLTK.

```python
#default sentence tokenizer
sentences = nltk.sent_tokenize(shakespeare)
sentences
```

```python
#using spacy
nlp = spacy.load("en_core_web_sm") #load english model
doc = nlp(shakespeare)
doc_sents = [sent.text for sent in doc.sents]
doc_sents
```
